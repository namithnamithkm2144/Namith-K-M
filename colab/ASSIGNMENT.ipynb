```python
print("===CROP SUITABILITY CHECK===")
soil_moisture=float(input("Enter soil moisture content(%)=",))
rainfall=float(input("enter rainfall(mm)=",))
if soil_moisture >=70 and rainfall >=200:
   print("suitable for rice")
else:
  print("not suitable for rice")
```

```python
print("===pest monitoring system===")
pest_count=1
while pest_count!=0 :
  pest_count=int(input("enter pest count:",))
  if pest_count!=0:
    print(f"pest count{pest_count}")

print("monitoring stopped")
```

```python
print("===ferilizer recommendation system===")
crop=input("Select crop(rice,wheat,maize)=",)
soil_nitrogen_level=float(input("enter soil nitrogen level=",))

if crop== "rice" and soil_nitrogen_level < 50:
  print("apply urea")
elif crop== "wheat" and soil_nitrogen_level<40:
  print("apply DAP")
else:
  print("fertilizer not required")
```

```python
print("===crop yield classification===")
crop_yield=float(input("Enter the yield(q)=",))
if crop_yield>=50:
    print("HIGH YIELD")
elif crop_yield>=30<=49 :
    print("MEDIUM YIELD")
else:
    print("LOW YIELD")
```

```python
print("====average rainfall analysis====")
rainfall=[]
for day in range(5):
  value=float(input(f"enter rainfall for day {day+1}: "))
  rainfall.append(value)

average_rainfall=sum(rainfall)/len(rainfall)
print(f"average rainfall={average_rainfall}")
```
