```python
import pandas as pd
data={
    'Department':['HR','Sales','IT','Sales','IT','HR','Sales'],
    'Employee':['A','B','C','D','E','F','G'],
    'Salary':[50000,60000,45000,48000,70000,72000,71000],
    'Experience':[2,4,3,5,6,7,5]
}
df=pd.DataFrame(data)
print(df)

```

```python
df.groupby(['Department','Experience']).sum()
```

```python
df.groupby('Experience')['Salary'].sum()
```

```python
df.groupby('Experience').filter(lambda x:len(x)>1)
```

```python
df.groupby('Department')['Salary'].transform('mean')
```

```python
df.groupby('Department').agg({'Salary':'sum','Experience':'mean','Employee':'sum'})
```

```python
df.groupby('Department')['Salary'].agg(['mean','std','sum'])
```

```python
df.groupby('Department')['Employee'].count()
```

```python
df.groupby('Department')['Salary'].mean()
```

```python
df.pivot_table(values='Experience',index='Department',columns='Salary',aggfunc='mean')
```

```python
import pandas as pd
data={
    'Crop':['maize','ragi','wheat','tomato','mango','citrus','palak'],
    'Type':['cereal','millet','cereal','vegetable','fruit','fruit','green vegetable'],
    'Yield':[500,150,400,600,400,600,100],
    'Price':[40,35,30,40,50,5,10],
    'Season':['annual','annual','annual','annual','perennial','perennial','annual']
}
df=pd.DataFrame(data)
print(df)
print('\n\n')
df1=df.groupby(['Type','Season']).sum()
print('\n',df1)
df2=df.groupby('Season').agg({'Yield':'sum','Price':'mean'})
print('\n\n',df2)
df[['Yield','Price']].plot()
```

```python
df.plot(kind='bar',x='Crop',y='Price')
```

```python
df3=df.sort_values('Yield')
print(df3)
```

```python
df.groupby('Season')['Yield'].transform('mean')
```
