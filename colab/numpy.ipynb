```python

import numpy as np
crop_yields=np.array([
    [12.3,33.3,43.4,34.4,12.5],
    [10.23,12.12,92.2,1.2,7.1],
    [12.2,12.23,21.4,34.3,34.2],
    [12.13,12.2,12,23.2,12]
])

print('Crop Yields Array(Tons):')
print(crop_yields)
print(f'\nShape:{crop_yields.shape}')
print(f'\nDataType: {crop_yields.dtype}')
print(f'\nTotal Elements:{crop_yields.size}')
print(f'\nNumber of dimensions:{crop_yields.ndim}')


total_yield=np.sum(crop_yields)
avg_yield_per_farm=np.mean(crop_yields,axis=1)
avg_yield_per_season= np.mean(crop_yields,axis=0)
max_yield=np.max(crop_yields)
min_yield=np.min(crop_yields)

print(f'\nTotal Production: {total_yield}Tons')
print(f'\nAverage per farm:{avg_yield_per_farm}Tons')
print(f'\nAverage per season:{avg_yield_per_season}Tons')
print(f'\nMaxium yield:{max_yield}Tons')
print(f'\nMinimum yield:{min_yield}Tons')


best_farm=np.argmax(avg_yield_per_farm)
print(f'\nBest Performing Farm:farm{best_farm+1}')
print(f'\nTheir average yield:{avg_yield_per_farm[best_farm]:.2f}Tons')
```

```python
import numpy as  np
yields=np.array([
    [45,48,52,49,53],
    [38,41,39,44,42],
    [52,55,58,56,60],
    [41,43,45,47,46]

])
weather_factors=np.array([1,0.95,1.05,0.98,1.02])
print('Original Yields(Tons/hectare):')
print(yields)
print('\nWeather Impact Factors:',weather_factors)
adjusted_yields=yields*weather_factors
print('\nWeather-Adjusted Yields')
print(adjusted_yields)
exceptional=adjusted_yields>55
print('\nExceptional yields(>55):')
print(adjusted_yields[exceptional])
categories=np.where(
    adjusted_yields>50,'high',
    np.where(adjusted_yields>40,'medium','low'))

print('\nYield Categories:')
print(categories)

selected_farms=[0,2]
print(f'\nSelected Farm Yields:')
print(yields[selected_farms])

good_years=adjusted_yields[(adjusted_yields>45)&(adjusted_yields<55)]
print(f'\nGood Performing Years(45-55Tons):')
print(good_years)


```
