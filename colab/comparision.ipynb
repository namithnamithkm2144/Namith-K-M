```python
x=20
y=15
print("x==y:",x==y)
print("x!=y:",x!=y)
print("x>y:",x>y)
print("x<y:",x<y)
print("x>=y:",x>=y)
print("x<=y:",x<=y)
```

```python
a=10
b=6
print("bitwise AND:",a&b)
print("bitwise OR:",a|b)
print("bitwise XOR:",a^b)
print("bitwise NOT:",~a)
print("leftshift:",a<<1)
print("rightshift:",a>>1)
```

```python
rainfall=True
soil_moisture=False
print("AND:",rainfall and soil_moisture)
print("OR:",rainfall or soil_moisture)
print("NOT:",not rainfall)
```
