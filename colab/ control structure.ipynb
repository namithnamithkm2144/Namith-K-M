```python
print("\n===crop recommendation system===")
rainfall=251 #mm average
soil_type="sandyloamy"

if rainfall>500 and soil_type=="clay" :
 crop="rice"
elif rainfall>250 and soil_type in["sandyloamy","loamy"]:
  crop="wheat"
elif rainfall<100 and soil_type== "red soil":
  crop="ragi"
else:
  crop="horti crops"
print(f"recommended crop:{crop}")
```

```python
print("\n====crop quality grading===")
weight=50
blemishes=1
colour_scale=9# out of 10

if weight>= 80 and blemishes ==0 and colour_scale>=8:
  grade="A"
elif weight>=60 and blemishes<=2 and colour_scale>=6:
  grade="B"
elif weight>=40:
  grade="C"
else:
  grade="D(reject)"
print(f"Crop Grade:{grade}")
```
