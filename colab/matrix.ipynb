```python
import numpy as np
matrix1=np.array([2])
matrix2=np.array([[2,3],[3,3]])
matrix3=np.array([[[2,3,4],[2,4,5],[4,5,7]]])
det=np.linalg.det(matrix3)
print("\n",matrix1)
print("\n",matrix2)
print("\n",matrix3)
print("\n",det)
```

```python
import numpy as np
matrix1=np.array([2])
matrix2=np.array([[2,3],[3,3]])
matrix3=np.array([[[2,3,4],[2,9,5],[4,5,8]]])
inverse=np.linalg.inv(matrix3)
print(inverse)
rank =(np.linalg.matrix_rank(matrix1))
print(rank)
```

```python
matrix3=np.array([[[2,3,4],[2,9,5],[4,5,8]]])
eigenvalues,eigenvectors=np.linalg.eig(matrix3)
print(eigenvalues)
print("\n",eigenvectors)
```
