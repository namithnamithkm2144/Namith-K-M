```python
def test_control(tup):
  count_even=0
  for val in tup:
    if val in tup:
      if val%2==0:
        count_even+=1
      elif val>2:
        break
  return count_even
test_control((1,2,3,4))
```

```python
days=[9,8,7,6]
window=[]
i=0
while i<len(days)-1:
  j=i+3
  while j<len(days):
    if days[j]-days[i]<=8:
      window.append((i,j))
    j+=2
  i+=1
print(window)

```
