```python
import pandas as pd
import numpy as np
df=pd.DataFrame({
    'Name':['Alice','Bob','Charlie','David','Eva','Frank'],
    'Department':['HR','IT','IT','Finance','HR','Finance'],
    'Salary':[60000,70000,70000,50000,52000,60000],
    'Experience':[2,5,4,6,3,6],

})
print(df)


```

```python
sal=df.sort_values('Salary')
print(sal)
```

```python
sal=df.sort_values(['Salary','Experience'], ascending=[False,True])
print(sal)
```

```python
dec=df.sort_values('Salary',ascending=False)
print(dec)
```

```python
idx=df.sort_index(ascending=False)
print(idx)
```

```python
df['rank_salary']=df['Salary'].rank()
print(df)
```

```python
df['rank_salary']=df['Salary'].rank(ascending=False)
print(df)
sor=df.sort_values('rank_salary')
print('\n',sor)
```

```python
df['dense_rank_Salary']=df['Salary'].rank(ascending=False,method='dense')
print(df)
```

```python
nam=df.sort_values('Name',ascending=False)
print(nam)
```

```python
df=df['Department'].value_counts()
print(df)
```
