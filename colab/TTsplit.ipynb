```python
import numpy as np
import pandas as pd
np.random.seed(42)
n=120
rainfall=np.random.uniform(400,1200,n)
fertilizer=np.random.uniform(50,300,n)
temperature=np.random.uniform(18,35,n)

yield_tons=(0.005*rainfall+0.03*fertilizer-0.1*temperature+np.random.normal(0,2,n))

data=pd.DataFrame({
      "Rainfall":rainfall,
      "Fertilizer":fertilizer,
      "Temperature":temperature,
      "yield":yield_tons
  })
print(data.head())
from sklearn.model_selection import train_test_split
x=data[["Rainfall","Fertilizer","Temperature"]]
y=data["yield"]
#val=validation
x_train,x_temp,y_train,y_temp=train_test_split(x,y,test_size=0.30,random_state=1)

x_val,x_test,y_val,y_test = train_test_split(x_temp,y_temp,test_size=0.50,random_state=1)
print("Training set:",x_train.shape)
print("Validation set:",x_val.shape)
print("Test set:",x_test.shape)

from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)

#pred=prediction
from sklearn.metrics import mean_absolute_error , mean_squared_error,r2_score
def evaluate_model(x,y,name):
  y_pred=model.predict(x)
  mae=mean_absolute_error(y,y_pred)
  mse=mean_squared_error(y,y_pred)
  rmse=np.sqrt(mse)
  r2=r2_score(y,y_pred)

  print(f"\n{name} Set Performance")
  print("MAE:",mae)
  print("MSE:",mse)
  print("RMSE:",rmse)
  print("R^:",r2)


evaluate_model(x_train,y_train,"Training")
evaluate_model(x_val,y_val,"Validation")
evaluate_model(x_test,y_test,"Test")


from sklearn.model_selection import cross_val_score

def cross_validation_results(k):
  scores=cross_val_score(model,x,y,scoring="neg_mean_squared_error",cv=k)
  rmse_scores=np.sqrt(-scores)

  print(f"\n{k}-Fold Cross validation Results")
  print("RMSE per fold:",rmse_scores)
  print("Average RMSE:",rmse_scores.mean())
  print("Std Dev RMSE:",rmse_scores.std())

for k in [3,5,10]:
  cross_validation_results(k)


new_data_values = np.array([[410,110,26]])
new_data_df = pd.DataFrame(new_data_values, columns=x_train.columns)
predicted_yield=model.predict(new_data_df)
print("Predicted Yield (tons/ha):",round(predicted_yield[0],2))
```
