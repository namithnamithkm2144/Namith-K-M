```python
crop_prices={
    "rice":40,
    "wheat":50,
    "maize":30
}
crop_yield={
    "rice":1239,
    "wheat":938,
    "maize":27
}


for crop,price,yield_kg in zip(
    crop_prices.keys(),
    crop_prices.values(),
    crop_yield.values()):
    print(f"{crop}: price={price} ,yield={yield_kg}")
```
