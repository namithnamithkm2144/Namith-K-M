```python
code=["crp1_rice_2930",
      "crp2_maize_2342",
      "frt3_mango_3475"
      ]
clean_codes=[]
for codes in code:
  if codes.startswith("crp"):
    clean_codes.append(codes)

print(clean_codes)
```

```python
for codes in code:
  if codes.startswith("crp"):
    code,crop,yield_kg=codes.split("_")
    print(crop,"crop with code",code,"produces yield of",yield_kg,"kg")
```
